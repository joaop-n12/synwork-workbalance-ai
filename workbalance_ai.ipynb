```python
from IPython.display import clear_output
import numpy as np
from time import sleep
from datetime import datetime
from statistics import mean
import random
```

```python
lista_colaboradores = []

for c in range(5):
    while True:
        try:
            clear_output()

            print(f'=== CADASTRO DO COLABORADOR {c+1}/5 ===\n')

            nome = str(input('Digite seu nome: ')).upper()
            departamento = str(input('Digite seu departamento: ')).upper()
            htd = float(input('Horas trabalhadas por dia: '))
            pausas = int(input('Pausas por dia (quantidade): '))

            while True:
                nivel_estresse = int(input('Nível de estresse (1 a 5): '))
                if 1 <= nivel_estresse <= 5:
                    break
                print("⚠ O estresse deve ser entre 1 e 5. Tente novamente.")

            tarefas_concluidas = int(input('Tarefas concluídas: '))

            print('\nAguarde...')
            sleep(2)
            print('Colaborador registrado com sucesso!')
            sleep(2)

            lista_colaboradores.append({
                'nome': nome,
                'departamento': departamento,
                'horas_trabalhadas': htd,
                'pausas': pausas,
                'estresse': nivel_estresse,
                'tarefas_concluidas': tarefas_concluidas
            })

            break

        except ValueError:
            print("\n❌ Valor inválido! Digite novamente apenas este colaborador.")
            sleep(2)


clear_output()
print('Colaboradores salvos com sucesso!')
sleep(2)
```

```python
def calcular_media_horas(lista):
  return mean(c['horas_trabalhadas'] for c in lista_colaboradores)

def maior_estresse(lista):
    return max(lista, key=lambda c: c['estresse'])['nome']

def colaboradores_produtivos(lista):
    return [c['nome'] for c in lista if c['tarefas_concluidas'] >= 5]

def alerta_equilibrio(lista):
    return [c['nome'] for c in lista if c['estresse'] >= 4 and c['pausas'] <= 1]

def feedback(colaborador):
    nome = colaborador['nome']
    estresse = colaborador['estresse']
    pausas = colaborador['pausas']
    tarefas = colaborador['tarefas_concluidas']

    if tarefas >= 5 and estresse <= 2:
        if pausas >= 2:
            return f'{nome}: excelente desempenho! Você mantém ritmo forte com bom equilíbrio.'
        else:
            return f'{nome}: muito produtivo.'

    if tarefas >= 5 and estresse >= 4:
        return f'{nome}: produtividade alta, mas o estresse está elevado. Considere reduzir o ritmo.'

    if tarefas <= 2 and estresse <= 2:
        return f'{nome}: você está tranquilo, mas precisa melhorar o foco para aumentar as entregas.'

    if tarefas <= 2 and 2 < estresse < 4:
        return f'{nome}: ritmo baixo e estresse moderado. Reorganize prioridades para melhorar.'

    if tarefas <= 2 and estresse >= 4:
        return f'{nome}: baixa entrega sob alto estresse. Faça pausas e respire.'

    if estresse >= 4 and pausas <= 1:
        return f'{nome}: alta carga e poucas pausas. Tente intercalar descansos.'

    if 2 < estresse < 4 and pausas <= 1:
        return f'{nome}: aumente suas pausas para evitar desgaste.'

    if 3 <= tarefas <= 4 and 2 <= estresse <= 3:
        return f'{nome}: bom desempenho, mantenha o equilíbrio!'

    if tarefas >= 4 and pausas >= 4:
        return f'{nome}: bom desempenho, mas muitas pausas podem afetar o fluxo.'

    mensagens_neutras = [
        f'{nome}: desempenho estável. Continue evoluindo.',
        f'{nome}: ritmo consistente! Pequenos ajustes podem ajudar.',
        f'{nome}: tudo dentro do esperado. Boa constância.'
    ]

    return random.choice(mensagens_neutras)
```

```python
print('RELATÓRIO WORKBALANCE AI')
print('-'*40)

horas = np.array([c['horas_trabalhadas'] for c in lista_colaboradores])
media_horas = np.mean(horas)
desvio_horas = np.std(horas)
estresse = np.array([c['estresse'] for c in lista_colaboradores])
media_estresse = np.mean(estresse)


print(f'\nMédia de horas trabalhandas: {media_horas:.1f}h')

print(f'Desvio padrão de horas: {desvio_horas:.1f}h')

print(f'Média de estresse: {media_estresse:.1f}')

print(f'Colaborador mais estressado: {maior_estresse(lista_colaboradores)}')

print(f'Colaboradores com 5+ tarefas: {colaboradores_produtivos(lista_colaboradores)}')

print(f'Alerta de equilíbrio: {alerta_equilibrio(lista_colaboradores)}\n\n')



print("\nFEEDBACK INDIVIDUAL")
print('-'*40)
for c in lista_colaboradores:
    print(feedback(c))

```

```python
def salvar_relatorio(lista_colaboradores):

    try:
        agora = datetime.now().strftime('%d/%m/%Y %H:%M:%S')

        horas = np.array([c['horas_trabalhadas'] for c in lista_colaboradores])
        media_horas = np.mean(horas)
        desvio_horas = np.std(horas)
        estresse = np.array([c['estresse'] for c in lista_colaboradores])
        media_estresse = np.mean(estresse)

        feedbacks = [feedback(c) for c in lista_colaboradores]

        texto = (
            'RELATÓRIO WORKBALANCE AI\n'
            f'{"-"*40}\n'
            f'Média de horas trabalhadas: {media_horas:.1f}h\n'
            f'Desvio padrão de horas: {desvio_horas:.1f}h\n'
            f'Média de estresse: {media_estresse:.1f}\n'
            f'Colaborador mais estressado: {maior_estresse(lista_colaboradores)}\n'
            f'Colaboradores com 5+ tarefas: {colaboradores_produtivos(lista_colaboradores)}\n'
            f'Alerta de equilíbrio: {alerta_equilibrio(lista_colaboradores)}\n\n'
            'FEEDBACK INDIVIDUAL\n'
            f'{"-"*40}\n' +
            '\n'.join(feedbacks) +
            '\n\n\n\n'
            f'Relatório gerado em: {agora}\n'
        )

        try:
            with open('relatorio_workbalance.txt', 'w', encoding='utf-8') as f:
                f.write(texto)
        except (OSError, IOError):
            return False

        return True

    except Exception:
        return False
```

```python
if salvar_relatorio(lista_colaboradores):
    print('✔ Relatório gerado com sucesso!')
    print('Arquivo salvo como: relatorio_workbalance.txt')
else:
    print('❌ Falha ao salvar o relatório. Verifique permissões ou espaço em disco.')
```
